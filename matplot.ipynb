Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
x=["part1","part2","part3","part4","part5"]
y=[78,90,67,98,69]
colors=['red','orange','black','green','yellow']
plt.xlabel("Parts of Salar movie")
plt.ylabel("Popularity of Salar movie in %")
plt.title("Popularity of  different parts of Salar movie",fontsize=15 )
plt.bar(x , y,color=colors)
plt.show()
```

from dataset importing data and making charts

```python
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_excel('/content/expense3.xlsx')
dataset=pd.DataFrame(data)
print(dataset.head())
group_by=dataset.groupby("Payment Mode")["Amount"].sum()
print(group_by)

plt.bar(group_by.index,group_by.values)
plt.show()
```

Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
x=["jan17","jan18","jan19","jan20","jan25"]
y=[100,500,60,90,600]
y1=[500,100,600,200,600]
plt.plot(x,y,marker='o',color='red',ls='-',label='year 2023')
plt.plot(x,y1,marker='D',color='blue',ls='--',label='year 2024')
plt.legend()

plt.show()
```

scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
import random
x=np.random.randint(1,10,50)
y=np.random.randint(1,100,50)
plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
x=np.random.randint(1,10,50)
y=np.random.randint(1,100,50)
colors=np.random.randint(1,100,50)


plt.scatter(x,y,marker="*", cmap="ocean",c=colors)
plt.colorbar()
plt.show()
```
