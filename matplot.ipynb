Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
x=["part1","part2","part3","part4","part5"]
y=[78,90,67,98,69]
colors=['red','orange','black','green','yellow']
plt.xlabel("Parts of Salar movie")
plt.ylabel("Popularity of Salar movie in %")
plt.title("Popularity of  different parts of Salar movie",fontsize=15 )
plt.bar(x , y,color=colors)
plt.show()
```

pieplot (pie chart)

```python
import matplotlib.pyplot as plt
brands=["Tesla","BMW","Toyota","Honda","Scorpio"]
x=[90,45,85,40,30]

plt.pie(x,labels=brands,autopct="%.2f")
plt.title("Market Value of different car brands")
plt.legend(brands)
plt.show()
```

Stack Plot

```python
import matplotlib.pyplot as plt
days=['sun','mon','tue','wed','thu','fry','sat']
firstWeek=[30,32,12,10,4,90,56]
secondWeek=[50,20,12,44,4,23,45]
thirdWeek=[30,56,7,86,56,44,23]
plt.stackplot(days,firstWeek,secondWeek,thirdWeek,labels=['firstWeek','secondWeek','thirdWeek'])
plt.legend()
plt.show()
```

Step Plot

```python
import matplotlib.pyplot as plt
days=['sun','mon','tue','wed','thu','fry','sat']
people=[30,32,12,10,4,90,56]
plt.step(days,people,where="mid",marker="o")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data= pd.read_excel('/content/expense3.xlsx')
dataframe=pd.DataFrame(data)

groupdata=dataframe.groupby("Category").agg({"Amount":"sum"})
print(groupdata)
plt.step(groupdata.index,groupdata["Amount"],marker="*")

plt.show()
```
