# Label Enocding

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Fit the label encoder on the 'Property_Area' column of the dataset
label_encoder.fit(dataset['Property_Area'])

```

# Handling outliers

```python
dataset.describe()
```

```python
sns.boxenplot(x='ApplicantIncome',data=dataset)
plt.show()
```

```python
sns.distplot(dataset['ApplicantIncome'])
plt.show()
```

**removing outliers uding IQR**

```python
q1=dataset['ApplicantIncome'].quantile(0.25)
q3=dataset['ApplicantIncome'].quantile(0.75)
IQR=q3-q1
IQR
```

```python
min_range=q1-(1.5*IQR)
max_range=q1+(1.5*IQR)
min_range,max_range
```

```python
new_dataset=dataset[dataset['ApplicantIncome']<=max_range]
```

```python
new_dataset.shape
```

```python
sns.boxenplot(x='ApplicantIncome',data=new_dataset)
plt.show()
```

**removing outliers uding Z - Score **

*   z= x-μ/σ




```python
sns.boxenplot(x='CoapplicantIncome',data=dataset)
plt.show()
```

```python
sns.distplot(dataset['CoapplicantIncome'])
plt.show()
```

```python
min_range=dataset['CoapplicantIncome'].mean() - (3*dataset['CoapplicantIncome'].std())
max_range=dataset['CoapplicantIncome'].mean() + (3*dataset['CoapplicantIncome'].std())
min_range,max_range
newdata = dataset[dataset['CoapplicantIncome'] <= max_range]
```

```python
z_score= (dataset['CoapplicantIncome'] - dataset['CoapplicantIncome'].mean()) / (dataset['CoapplicantIncome'].std())
```

```python
dataset['z_score']=z_score
```

```python
dataset[dataset['z_score']<3].shape
```

```python
sns.boxenplot(x='z_score',data=dataset)
plt.show()
```
