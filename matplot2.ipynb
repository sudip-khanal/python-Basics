subplot example

```python
import matplotlib.pyplot as plt

x=[48,34,23,5,4,34,45,45]
y=[66,45,34,23,65,43,33,45]
plt.subplot(1,2,1)
plt.plot(x,y)

x=[54,35,66,53,42,35,45]
y=[66,45,34,23,65,43,33]
plt.subplot(1,2,2)
plt.suptitle("Subtitles")
plt.plot(x,y)
plt.savefig("chart.png")
plt.show()
```
